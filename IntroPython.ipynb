# Declaración de Funciones

Para declarar una función se debe indicar cuales son sus parametros y la salida o resultado de la funcion


```python
def factorial(i):
    factorial=1;
    for i in range(1,i+1):
        factorial*=i;
    return factorial;

print factorial(2)
b=factorial(5);
print 'b= 1x2x3x4x5 =',b
```

# Hacer Gráficos con Matplotlib

```python
#%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt


x= np.arange(1,10,1);
y = np.arange(1,10,1);
##fig = plt.figure()
plt.plot(x,y,"ro-", label="Recta", linewidth=2, markersize=8)
##plt.show()

```

## 2.1 La ecuación de Schrödinger

La ecuación que determina la evolución de un sistema cuántico es la ecuación de Schrödinger, esta describe la evolución de la función de onda de un sistema de partículas. Donde el Hamiltoniano pasa de ser una función real a un operador en el espacio de Hilbert determinado por el mismo $\hat{H}=\hat{T}+\hat{V}$ 

$$ i\hbar {\frac {d}{dt}}\vert \Psi (t)\rangle ={\hat {H}}\vert \Psi (t)\rangle $$

La ecuación de Schrödinger determina la dependencia temporal de forma explicita en representación en el espacio real:  

$$ i\hbar {\frac {\partial }{\partial t}}\Psi (\mathbf {r} ,t)=\left[{-\frac {\hbar ^{2}}{2m}}\nabla ^{2}+V(\mathbf {r} ,t)\right]\Psi (\mathbf {r} ,t)$$

Observamos que el operador energía cinética es $$\hat{T}=-\frac{\hbar ^{2}}{2m}\nabla ^{2}$$ para efectos de este curso estamos interesados en el caso estacionario de la ecuación de Schrödinger 

$$\Psi (\mathbf {r} ,t)=\Phi(t)\psi(\mathbf {r})$$

Considerando adicionalmente $V(r)$ un potencial independiente del tiempo, tenemos:

$$ i\hbar {\frac {\partial }{\partial t}}\Phi(t)\psi(\mathbf {r})=\left[{-\frac {\hbar ^{2}}{2m}}\nabla^{2}+V(\mathbf {r})\right]\Phi(t)\psi(\mathbf {r})$$

Usando el Ansatz $\Phi(t)=\exp\left(-\frac{iEt}{\hbar}\right)$ la ecuación es simplificada: 

$$ E \psi(\mathbf {r})=\left[{-\frac {\hbar ^{2}}{2m}}\nabla^{2}+V(\mathbf {r})\right]\psi(\mathbf {r})$$


## 2.2 El Principio Variacional 

Supongan tenemos la ecuación de Scrhodinger indpendiente del tiempo, para una dada función de onda $\psi$

$$H\psi =E\psi$$

Con H un hamiltoniano de un sistema de muchos cuerpos bastante complejo, si $\psi$ es normalizada el __Principio Variacional__ nos dice que el valor esperado de energía de la función de onda $\psi$  es menor que la energía del estado base $E_{0}$

$$ E_{0}\leq \langle \psi | H | \psi \rangle  $$

Esto quiere decir que si hacemos una variación de $\psi$ el valor esperado de H puede ser minimizado para obtener una aproximación razonable al estado Base. Para probar el principio variacional partamos del hecho que debe existir un conjunto de autovectores y autovalores para H:

$$H|\psi_n\rangle =E_n|\psi_n\rangle$$

Por definición estos autovectores son ortonormales 

$$\langle \psi_n|\psi_m \rangle=\delta_{mn}$$

Sus autovalores estan ordenados: 

$$E_{0}<E_{1}<E_{2}<\cdots<E_{n-1}<E_{n}$$

De tal forma que este conjunto de autovectores forman una base para representar cualquier función de onda $\psi$, como una combinación lineal de estos:

$$|\psi\rangle=\sum_{n} C_{n}|\psi_{n}\rangle$$

Ya que $\psi$ es normalizada, los coeficientes cumplen la condición:

$$\sum_{n}|C_{n}|^{2}=1$$

Ahora calculemos el valor esperado de $\psi$ en terminos de  esta expansión: 

$$
\begin{align}
\langle \psi | H | \psi \rangle & =  \sum_{m} C^{*}_{m}\langle \psi_m | H \sum_{n} C_{n}|\psi_n\rangle\\
& =  \sum_{m,n} C^{*}_{m} C_{n}|\langle \psi_m |H|\psi_n\rangle\\
& =  \sum_{m,n} C^{*}_{m} C_{n}| E_{n}\langle psi_m |\psi_n\rangle \\
& =  \sum_{m,n} C^{*}_{m} C_{n}| E_{n}\delta_{mn} \\
& =  \sum_{n} |C_{n}|^{2} | E_{n}\\
\end{align}
$$

Esta expresión puede ser reordenada:


$$
\langle \psi | H | \psi \rangle  = |C_{0}|^{2} | E_{0}+ \sum_{n>0} |C_{n}|^{2}  E_{n} 
$$

Adicionalmente la condición de normalización de $\psi$ sobre los coeficientes:

$$|C_{0}|^{2}= 1-\sum_{n>0}|C_{n}|^{2}$$

Reemplazando esto en la expresión anterior:

$$
\begin{align}
\langle \psi | H | \psi \rangle  &= \left( 1-\sum_{n>0}|C_{n}|^{2}\right)| E_{0}+ \sum_{n>0} |C_{n}|^{2} E_{n}\\
&= E_{0}+ \sum_{n>0} |C_{n}|^{2} (E_{n}-E_{0})\\
\end{align}
$$

Esta expersión definida positivamente ya que $$(E_{n}-E_{0})>0\; \forall n>0 $$ de esta manera, obetenemos el __Principio Variacional__


$$\langle \psi | H | \psi \rangle\geq E_{0}$$

Luego siempre podemos encontrar una función $\psi$ que minimizada sera un limite superior al estado base. El exito del metodo Variacional dependera de la parametrización de la función de onda y de su minimizacion. Este principio constituye un pilar importante de la mayoria de metodos de Monte Carlo cuántco, tambien lo es para otros metodos variacionales como Densidad funciónal y otros metodos de quimica cuántca.




## 2.3 Conectando el principio Variacional con Monte Carlo

La evaluación de un observable cuántco a temperatura cero, dada una función de onda $|\psi\rangle$ que define el estado del sistema: 

$$\langle \hat{O} \rangle = \langle \psi | \hat{O} | \psi\rangle$$

Usando la condición de normalización 

$$\langle \hat{O} \rangle = \frac{\langle \psi | \hat{O} | \psi\rangle}{\langle \psi | \psi\rangle}$$

Supongamos usamos una base de estados en el espacio de Hilbert que es completa y ortornormal tal que: 

$$\sum_{N}|x\rangle\langle x|=\mathbb{1}\$$

Usando esta relación de completes en el denominador y numerador de la expresión del observable:

$$
\langle \hat{O} \rangle = \frac{\sum_{N}\langle \psi |x\rangle\langle x| \hat{O}\psi\rangle}{\sum_{N} \langle \psi |x\rangle\langle x| \psi\rangle}
$$

Si multiplicamos por $1=\frac{\langle x| \psi\rangle}{\langle x| \psi\rangle}$:

$$\begin{align}
\langle \hat{O} \rangle &= \frac{\sum_{N}|\langle x| \psi\rangle|^2 \frac{\langle x| \hat{O} |\psi\rangle}{\langle x| \psi\rangle}}{\sum_{N} \langle \psi |x\rangle\langle x| \psi\rangle}\\
& =\frac{\sum_{N}|\psi(x)|^2  O_{L}(x)}{\sum_{N} |\psi(x)|^2}
\end{align}
$$

Con el estimador del operador local: 

$$O_{L}(x)=\frac{\langle x| \hat{O} |\psi\rangle}{\langle x| \psi\rangle}$$

Para hacer la conexión con Monte Carlo el observable debe ser la forma que es el promedio del observable por la probabilidad, podemos identificar:  

$$P(x)=\frac{\psi(x)|^2}{\sum_{N} |\psi(x)|^2}$$

Esta puede ser interpretada como probabilidad ya que es siempre positiva para todo x, además es fácil de observar que:

$$\sum_{N}P(x)=1$$


El observable local es la función sobre la cual se calcula el valor esperado con probabilidad $P(x)$, reemplazando el observable por el hamiltoniano H: 

$$E_{L}(x)=\frac{\langle x| \hat{H} |\psi\rangle}{\langle x| \psi\rangle}$$

Luego es posible definir un algoritmo que genere diferentes configuraciones ${|x_{n}\rangle}$, una cadena de Markov usando el algoritmo de Metropolis, que muestrea $P(x)$. Asi el valor esperado del observable puede ser aproximado como el promedio del observable local sobre diferentes configuraciones: 

$$
\langle \hat{O} \rangle =\frac{1}{N}\sum_{n=1}^{N} O_{L}(x_{n})
$$


Para la energía total del sistema $E(\psi(\mathbf(x))$ o $E_{VMC}$


$$
E = \frac{1}{N}\sum_{n=1}^{N} E_{L}(x_{n})
$$

## 2.3 Principio de Varianza Cero

La varianza de $\hat{H}$ de manera analoga es es definida como:  

$$\begin{align}
\sigma_{E}^2 &=\frac{\langle \psi | (\hat{H}-\langle H \rangle)^2 |\psi \rangle }{\langle \psi|\psi\rangle }\\
&=\frac{\langle \psi | (\hat{H}-E)^2 |\psi \rangle }{\langle \psi|\psi\rangle }\\
&=  \left(\frac{\langle \psi | \hat{H}^2|\psi \rangle }{\langle \psi|\psi\rangle}-E^2\right)    
\end{align}
$$

Para obtener un buen estimado de la energía la varianza debe ser minimizada, el primer termino de la varianza es
un promedio del valor cuadrado de la energía local:

$$  
\frac{\langle \psi | \hat{H}^2|\psi \rangle }{\langle \psi|\psi\rangle}=\frac{\sum_{N} |\psi(\mathbf{x})|^2 |E_{L}(\mathbf(x))|^2}{\sum_{N}|\psi(\mathbf(x))|^2}
$$


En el caso especial en que la función de onda variacional $\psi(\mathbf(x)$ es un autestado de $H$ tenemos

$$
\begin{align}
E_{L}(x)&=\frac{\langle x| \hat{H} |\psi\rangle}{\langle x| \psi\rangle}\\
&=E_{n}\frac{\langle x|\psi\rangle}{\langle x| \psi\rangle}\\
&=E_{n} 
\end{align}
$$

La energía local $E_{L}(x)=E_{n}=Constante$, es decir independiente de la configuración $\mathbf{x}$, es facil observar que:  

$$
\sigma_{E}^2 = 0
$$

En general nuestro objetivo es minimizar la varianza de la energía local $E_{L}(x)$ tal que nuestra función de onda variacional se acerca al estado base con gran presicion.

Usualmente $\psi(\mathbf(x))=\psi(\mathbf(x),\{\alpha_i\})$, donde $\left\{\alpha_1,\alpha_2,\cdots\alpha_n\right\}$ son un conjunto de parametros ajustables. De tal manera que son escogidos para minimizar la varianza, acercando la función de prueba al estado base lo mejor posible.  



## 2.3.1 Algoritmo Metropolis

El algoritmo Monte Carlo seria el siguiente is $\mathbf{X}$ es una configuración del sistema descrito por el hamiltoniano H, 

1. Generar una configuración  $\mathbf{X'}\rightarrow\mathbf{\zeta}+\mathbf{X}$
2. Calcular la razón del criterio de aceptación 

$$p_{accept}(\mathbf{X'}|\mathbf{X})= Min\left\{1,\frac{T(\mathbf{X}|\mathbf{X'})\frac{|\psi(\mathbf{X'})|^2}{\sum_{N} |\psi(\mathbf{X''})|^2}}{T(\mathbf{X'}|\mathbf{X})\frac{|\psi(\mathbf{X})|^2}{\sum_{N} |\psi(\mathbf{X''})|^2}}\right\}$$
Simplificando 

$$p_{accept}(\mathbf{X'}|\mathbf{X})= Min\left\{1,\frac{T(\mathbf{X}|\mathbf{X'})|\psi(\mathbf{X'})|^2}{T(\mathbf{X'}|\mathbf{X})|\psi(\mathbf{X})|^2}\right\}$$

3. Aceptar o rechazar el movimiento propuesto 

4. Calcular la energía local $E_{L}$ sumarla para estimar $E$

Luego se repiten los pasos, sucesivamente generando la candena de markov de estados que muestrean $P(\mathbf{X})$ y satisfacen el principio variacional. 

La idea es que Monte Carlo nos da la mejor energía variacional para una función de onda $\psi(x,\left\{\mathbf{\alpha}\right\})$, con  ${\mathbf{\alpha}\right\}$ un conjunto de parametros de los cuales depende la función de onda. La ida es optener los parametros optimos para que la energia Variacional sea lo más cercana posible a la energía del estado base.  


## 2.3.2 Monte Carlo Variacional: El Oscilador Armonico 1d

El Oscilador armonico esta dado por el siguiente hamiltoniano en una dimensión:

$$
-\frac{\hbar}{2m} \frac{d^2}{dx^2}+\frac{1}{2}\omega x^2
$$


Supongamos no conocemos la solución exacta a este problema queremos usar la función de onda de la siguiente forma:

$$
\psi_{T}(x,\alpha)= \exp(-\alpha x^2)
$$

Usando unidades normalizadas $m=\hbar=\omega=1$, antes de escribir nuestro algoritmo en python lo primero que debemos hacer es encontrar la función de la energía local data nuestra función de onda de prueba:

$$
\begin{align}
E_{L}(x)&=\frac{ \hat{H}\psi_{T}(x) }{ \psi_{T}(x)}\\
&=\frac{1}{\exp(-\alpha x^2)}\left(-\frac{1}{2}\frac{d^2}{dx^2}+x^2\right)\exp(-\alpha x^2)
\end{align}
$$

La segunda derivada de $\psi_{T}$:

$$
\frac{d^2}{dx^2} \exp(-\alpha x^2) =-2\alpha \exp(-\alpha x^2)+4\alpha^2 x^2 \exp(-\alpha x^2)
$$

Luego

$$
\begin{align}
E_{L}(x)&=\frac{1}{\exp(-\alpha x^2)}\left(-\frac{1}{2}\frac{d^2}{dx^2} \exp(-\alpha x^2)+\frac{1}{2}x^2\right)\\
&=\alpha -2\alpha^2 x^2 +\frac{1}{2}x^2\\
&=\alpha+(\frac{1}{2}-2\alpha^2) x^2
\end{align}
$$

```python
#Cargue desde el inicio las bibliotecas necesarias para no llamarlas de nuevo a lo largo del cuaderno jupyter. 
import numpy as np
#%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
#defina una funcion para E local
def E_loc(x, alpha):
    return alpha + x**2 * (0.5 - 2.0 * alpha**2);

#defina una funcion para la probabilidad de transicion
def dP(x_old, x_new, alpha):
    return np.exp(-2*alpha*(x_new**2-x_old**2));

#inicialize las variables del algorithm
delta = 1.0;
npasos=60000; 
alfa = 0.3; 
#vectores de numeros aleatorios
dx=delta*(np.random.uniform(-0.5,0.5,npasos));
#print dx
eta=np.random.uniform(0,1,npasos);
naccept=0;
#condicion inial
x=0.1;
E=np.zeros(npasos);
EC=np.zeros(npasos);
Eit=0.0;
for i in range(0, npasos):
    #mueva la particula a una nueva posicion
    xnew=x+dx[i];
     #Rechaza o acepta el movimiento
    if (eta[i]<=dP(x,xnew,alfa)):
        x=xnew;
        naccept+=1;
    E[i]=E_loc(x,alfa);    
    Eit+=E[i];
    EC[i]=Eit*(1.0/(i+1));

Ev=(1.0*Eit)/npasos;        
#dibuje el vector de energia local        
plt.figure()
plt.plot(EC, linestyle='solid', marker='o', markerfacecolor='r', markeredgecolor='r', markersize=1)
plt.xlabel("n pasos")
plt.ylabel("$E_{v}(npasos)$")
print "Energia Variacional Calculada", Ev    
```

## 2.4 Calculando el error en Monte Carlo cuando los datos estan correlacionados?

El error estadar de variables estocasticas independientes esta definido como: 

$$
\begin{align}
\sigma_{E}& =\sqrt{\frac{\sigma^2}{n}}\\
& =\sqrt{\frac{\langle E^2 \rangle -\langle E\rangle^2}{n}}\\
& =\sqrt{\frac{1}{n}\left({\overline{E^2} -\overline{E}^2}\right)}
\end{align}
$$

Dado que en en Monte Carlo la cadena de Markov generada dependen de la posición anterior vemos que hay una correlación es decir para un $t$ y $t'$ arbitratios: $x_{t}$ y $x_{t'}$ no son independientes $\langle x_{t} x_{t'}\rangle \neq 0$. Sin embargo existe un número de pasos "n_{c}" para los cuales $x_{t}$ y $x_{t'}$ son independientes $\langle x_{t} x_{t'}\rangle = 0$. Es deseable que dado $T(x_{t'} | x_{t})$ el tiempo de correlación se a corto de tal manera que los estados generados son independientes y muestrea con mas facilidad el espacio de configuración que define la distribución de equilibrio. 

Luego para calcular el error correctamente es deseable tener una representación de nuestra información de forma que no hay correlación en esa representation de nuestros para poder aplicar la definición de error para variables independientes. Una simple técnica como __análisis de bloque__ hacer el promedio sobre bloques de datos de la simulación por ejemplo si $E_{i}=E_{x_{i}}\; , \left\{E_{1},E_{2},\cdots, E_{npasos}\right\}$, podemos agrupar la información en promedios de bloques continuos de datos de tamaño  $t_{b}$ es decir los nuevos datos seria:

$$E({b_{i}})=\overline{E}_{b_{i}}=\frac{1}{t_b}\sum_{i\in t_b}E_{i}\;,\;\;\left\{\overline{E}_{1},\overline{E}_{2},\cdots, \overline{E}_{NB}\right\}  $$


Con $NB$ el número de bloques dado 

$$NB=n_{pasos}\bmod t_b$$

luego nuestros nuevos datos empezaran a verse como variables independientes. Básicamente lo que ocurre es que se hace el análisis para encontrar el tamaño en que el promedio de los bloques de datos estan distribuidos sin correlación.  

Ahora queda una pregunta mas, como cálculo $t_{b}$ el tamaño del bloque adecuado. Para esto lo mas simple es observar el cambio del cálculo del error con el número de bloque y luego escogemos el valor donde 

$$\frac{\partial \sigma_{E}(t_b)}{\partial t_b}=0$$.

## 2.5 El átomo de Helio

Ahora nos vamos a concentrar en la tarea de hacer el calculo de Monte Carlo variacional del átomo de Helio, para esto utilizamos el hamiltoniano del átomo de Helio en la arpximación de Born-Oppenheinmer, es decir tratamos el nucleo del átomo en una posición fija y nos enfocamos en la interacción electronica.  Usando unidades atomicas:

$$\hbar\; =\; e\; =\; m\; = 1 $$

$$-\frac{1}{2} (\nabla_1^2 + \nabla_2^2) - \frac{2}{r_1} - \frac{2}{r_2} + \frac{1}{r_{12}}$$

Los dos primeros terminos correponden a la energía cinetica, el sistema de de referencia esta centrado en el nucleo del átomo, luego el tercer y cuarto termino corresponen a la interación de coulmb con el nucle y el ultimo termino a la repulsiónentre los electrones.Ya que estamos interesados en la función de onda de los electrones, estas particulas al ser fermiones deben tener una función de onda antisimetrica. Para esto vamos a utilizar una función de onda antisimetrica: 


$$\psi(r_{1},r_{2})=\phi(r_1)\phi(r_2) \frac{1}{2}\left(|\uparrow\downarrow\rangle +|\downarrow \uparrow\rangle\right)$$


Donde hemos dejado la parte antisimetrica en las variables de espin, $r_i$,$r_ij$ son las distancias entre cada electron y el nucleo, asi como la distancia entre los dos electrones respectivamente. La ocupación de spin para este ejemplo vamos a dejarla fija, para que capture la antisimetria de la función de onda. Los orbitales de la función de onda deben  ser una buna aproximación que refleje al fisica del sistema. Para esto usualmente se sigue las restriccciones impuestas por __"Cusp Conditions"__ o condiciones de Cuspide, es decir las condiciones para eleminar o tratar con las singularidades entre la interacción de Coulumb cuando el nucleo y un electron estan cerca, o los electrones estan cerca de tal modo que $|r|\to 0$.

Suponga que los electrones en el átomo de Helio estan separadados y el segundo electron a distancia $r_2$, esta alejado del nucle, sin embargo el primer electron esta cerca $|r_1|\to 0$. La Función de energía local debe evadir esta divergencia, entonces debemos encontrar una función de onda que satisfaga esta condicion.  La energía local estaria dada por: 

$$E_{L}=\frac{1}{\psi_t}\left(-\frac{1}{2} \nabla_1^2 - \frac{2}{r_1}+E(r_{12},r_2)\right)\psi_t$$

Donde los terminos dominantes estan dados por $r_1$ y la contribución de $E(r_{12},r_2)$ puede ser considerada despreciable, escribiendo la energía cinetica en coordenadas esfericas, con momento algular $l=04 para el electron 1:

$$E_{L}=\frac{1}{R_t(r_1)}\left(-\frac{1}{2}\frac{d^2}{dr_{1}^{2}}-\frac{1}{r_1}\frac{d}{dr_1} - \frac{2}{r_1}+E(r_{12},r_2)\right)R_t(r_1)$$

Luego la condición de cuspide me pide eliminar los terminos divergentes $\frac{1}{r_1}$ esto es:

$$\lim_{r_1\to 0} E_L(R_t(r1))= \frac{1}{R_t(r_1)}\left(-\frac{1}{r_1}\frac{d}{dr_1} - \frac{2}{r_1}\right)R_t(r_1)$$

La segunda derivada no diverge entonces para mantener la energía local finita se debe cumplir:


$$\frac{1}{R_t(r_1)}\frac{d R_t(r_1)}{dr_1} =-2$$

En general es igual a Z la carga del nucle atomico, para Helio Z=2. esta ecuación implica que: 

$$R_T(r_1)\propto\exp(-Zr_1)$$


De manera analoga usando el mismo razonamiento para el electron 2, cuando $r_2\to 0$ y momento angular $l>1$

$$\frac{1}{R_t(r_2)}\frac{d R_t(r_2)}{dr_2} =-\frac{Z}{l+1}$$

Luego

$$R_T(r_2)\propto\exp\left(-\frac{Z}{l+1}r_2\right)$$


Otra restricción puede ser encontrada sobre la función de onda en el caso cuando los dos electrones estan muy cerca $r_{12}\to 0$, la ecuación radial es similar a la que aparece entre un electron y el nucleo cuando se acercan, despreciando la contribución del centro de masa de los electrones de manera similar a las aproximaciones anteriores: 


$$\lim_{r_{12}\to 0} E_L(R_t(r_{12}))= \frac{1}{R_t(r_{12})}\left(2\frac{d^2}{dr_{12}^{2}}+\frac{4}{r_{12}}\frac{d}{dr_{12}} - \frac{2}{r_{12}}-\frac{l(l+1)}{r_{12}^2}+E(r_{1},r_2)\right)R_t(r_{12})$$

Eliminando los terminos de la divergencia aparece la condicion:

$$
\frac{d R_t(r_{12})}{dr_{12}}=-\frac{1}{2(l+1)}R_t(r_{12})
$$

Luego considerando espines anti paralelos y l=0:

$$
R_T(r_{12})\propto \exp\left(\frac{r_{12}}{2}\right)
$$

En general estas condiciones pueden ser escritas en una forma mas general, la función de onda de prueba tiene la forma:

$$
\psi_{T}(R)=phi(r_{1})phi(r_{2})\cdots \phi(r_{N})\prod_{i<j}f(r_{ij})
$$


Las condiciones de cuspide nos llevaron a encontrar la forma funcional basica que define la mayor parte de modelos de funciones de onda para solidos y moleculas. La parte de las funciones $\phi(r_i)$ es considerada como la parte fermionica de la función de onda. Las funciones $f(r_{ij}) es la parte que adiciona correlación al sistema. 

Retomando el caso del átomo de Helio una función de onda que evade las condiciones de cuspide de los electrones con el nucleo seria:

$$
\begin{align}
\phi_1(r1) & =\exp(-\alpha r_1)\\
\phi_2(r2) & =\exp(-\alpha r_2)\\
\psi(r_1,r_2) & = \phi_1(r_1) \phi_2(r_2) = \exp(-\alpha (r_1+r_2))
\end{align}
$$

Donde $\alpha$ podria ser interpretado como una carga variacional efectiva. Considerando esta función de onda, para escribir el algorithmo Metropolis debemos conocer la expresiónde la energía local, en este caso corresponde a: 

$$
E_{L}(r_1,r_2)=-\alpha^2+\frac{\alpha}{r_{1}}+\frac{\alpha}{r_2}-\frac{2}{r_1}-\frac{2}{r_2}+\frac{1}{r_{12}}
$$

Luego podemos implementar el algoritmo usando las funciones que ya habiamos escrito para el oscilador armonico con pequenos cambios en la dimensionalidad del problema. En las unidades dadas el valor de refencia experimental de la energía del estado base del Helio es:

$$2.9037\;\; Ha$$

1. Escriba una función de energía local para esta función de onda de prueba. 

2. Escriba una función que calcule el radio de la probabilidad de la función de onda de prueba 

3. Escriba una función Metropolis para este sistema

4. haga una variaciaon de parametros para encontrar el $\alpha$ optimo y la energía variacional del mismo


```python
def E_locHev1(r_in,alpha):
    #pasar un arreglo o matriz de numpy como copia de objeto, necesario cuando la funcion tiene mas de un argumento
    r=r_in.copy();  
    r1=np.linalg.norm(r[0,:]);
    r2=np.linalg.norm(r[1,:]);
    r12=np.linalg.norm(r[0,:]-r[1,:]);
    Eloc=-(alpha**2)+(alpha/r1)+(alpha/r2)-(2.0/r1)-(2.0/r2)+(1.0/r12);
    
    return Eloc;


def dPHev1(r_in,rnew_in,alpha):
    r=r_in.copy();
    rnew=rnew_in.copy();
    r1=np.linalg.norm(r[0,:]);
    r2=np.linalg.norm(r[1,:]);
    rnew1=np.linalg.norm(rnew[0,:]);
    rnew2=np.linalg.norm(rnew[1,:]);
    dp=np.exp(-2.0*alpha*((rnew1+rnew2)-(r1+r2)));
    
    return dp;
    
    

def MetropolisHev1(alfa):
    #inicialize las variables del algorithm
    delta = 1.0;
    npasos=50000; 
    #vectores de numeros aleatorios
    dr=delta*(np.random.uniform(-0.5,0.5,(npasos,2,3)));
    #print dx
    eta=np.random.uniform(0,1,npasos);
    E=np.zeros(npasos);
    naccept=0;
    #condicion inial
    r=np.zeros((2,3));          
    r[0,1]=0.3;
    r[1,1]=-0.3;          
    Ev=0.0;
    for i in range(0, npasos):
        #mueva la particula a una nueva posicion
        rnew=r+dr[i,:,:];
        #Rechaza o acepta el movimiento
        if (eta[i]<=dPHev1(r,rnew,alfa)):
            r=rnew;
            naccept+=1;
            #print naccept
        E[i]=E_locHev1(r,alfa)
        Ev+=E[i];
    Ev=(1.0*Ev)/npasos;    
    err = blockerror(E);
    
    return Ev, err;


alfas=np.arange(0.1,5.0,0.1);

#print alfas.shape[0]
dataHev1=np.zeros((alfas.shape[0],2));

#print dataHev1.shape

for i in range(0,alfas.shape[0]):
    dataHev1[i,:]=MetropolisHev1(alfas[i]);

plt.figure()
plt.plot(alfas,dataHev1[:,0],'r*-')

iemin=int(np.argmin(dataHev1[:,0]));

print "indice valor minimo", iemin, "valor de alfa", alfas[iemin], dataHev1[iemin,0]
print "finish"
```
